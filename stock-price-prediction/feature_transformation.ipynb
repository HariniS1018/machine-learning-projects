**1. Import Necessary Libraries**

```python
from sklearn.preprocessing import PowerTransformer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

**2. To read the data into Python.**

```python
df = pd.read_csv('ds/cleaned_nifty50.csv')
df.head()
```

**3. Handling Outliers of Positively Skewed Attributes**

```python
positively_skewed_outlier_columns = ['Shares Traded', 'Turnover (Cr Rs)']
```

**3a. Log Transformation**

```python
nrows = math.ceil(len(positively_skewed_outlier_columns) / 2)
plt.figure(figsize=(14, nrows * 4))

for i, col in enumerate(positively_skewed_outlier_columns):
    log_transformed = np.log1p(df[col])

    plt.subplot(nrows, 2, i + 1)
    sns.boxplot(y=log_transformed)
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

**3b. Box-Cox Transformation**

```python
pt_boxcox = PowerTransformer(method='box-cox')

nrows = math.ceil(len(positively_skewed_outlier_columns) / 2)
plt.figure(figsize=(14, nrows * 4))

for i, col in enumerate(positively_skewed_outlier_columns):
    box_cox = pt_boxcox.fit_transform(df[[col]])

    plt.subplot(nrows, 2, i + 1)
    sns.boxplot(y=box_cox.reshape(-1))
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

**3c. Yeo-Johnson Transformation**

```python
pt_boxcox = PowerTransformer(method='yeo-johnson')

nrows = math.ceil(len(positively_skewed_outlier_columns) / 2)
plt.figure(figsize=(14, nrows * 4))

for i, col in enumerate(positively_skewed_outlier_columns):
    df[col] = pt_boxcox.fit_transform(df[[col]])

    plt.subplot(nrows, 2, i + 1)
    sns.boxplot(y=df[col].values)
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

```python
outlier_columns = ['Shares Traded', 'Turnover (Cr Rs)']

for col in outlier_columns:
    Q1= df[col].quantile(0.25)
    Q3= df[col].quantile(0.75)
    IQR=Q3-Q1
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR
    median_value = df[col].median()

    df[col] = df[col].astype('float64').copy()

    df.loc[df[col] > upper_bound, col] = upper_bound
    df.loc[df[col] < lower_bound, col] = lower_bound

    fig, axes = plt.subplots(1, 2, figsize=(14, 4))

    sns.boxplot(x=df[col], ax=axes[0])
    axes[0].set_title(f'Boxplot of {col}')

    sns.histplot(df[col], kde=True, bins=30, ax=axes[1])
    axes[1].set_title(f'Distribution of {col}')
    axes[1].set_xlabel(col)
    axes[1].set_ylabel('Frequency')

    plt.tight_layout()
    plt.show()
```

```python
print("\n--- Bivariate Analysis: Numerical vs. Target Variable ---")
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()

nrows = math.ceil(len(numerical_cols) / 2)
plt.figure(figsize=(12, nrows * 5))

for i, x_col in enumerate(numerical_cols):
    if x_col == 'Close':
        continue
    plt.subplot(nrows, 2, i + 1)
    sns.scatterplot(x=x_col, y='Close', data=df, hue='Close', palette='plasma', alpha=0.8)
    plt.title(x_col + " vs. Close Price")
    plt.xlabel(x_col)
    plt.ylabel('Close Price')
    plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

**Conclusion of Feature Transformation**,

After experimenting with Log Transformation, Box-Cox Transformation and Yeo-Johnson methods, none of the methods were good in transforming attributes `Shares Traded` or `Turnover (Cr Rs)`.
